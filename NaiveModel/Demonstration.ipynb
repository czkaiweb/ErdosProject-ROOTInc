# Naive Bidding Strategy

Here we explore a very trivial strategy, which doesn't work too well. The idea is simple, we group the data by unique combinations of the 4 features (Currently insured, Number of vehicles, Number of drivers,Marital status) and then look at the distribution of successful bids over the rank in the search.

Then the bidding strategy is simply, given a feature with its rank, create a random variable that bids 10 with a probability equal to that of successful bid over a that rank.

An additional option we examine is allowing the distributions to be updated with each bid. This allows the distributions to adjust to changes in the market. Unfortunately the data that we have doesn't allow us to extend the model to also factor in bid-rank distributions, which could be one way to improve perfomance.

We assess the performance of the strategy by performing a 5-fold cross validation and look at the average reward.

```python
from naive_bidder import naive_bidder
import pandas as pd
from sklearn.model_selection import train_test_split,StratifiedKFold
import matplotlib.pyplot as plt
import numpy as np
```

```python
root_ins_df = pd.read_csv('../DataExploration/Root_Insurance_data.csv')
train_data, test_data = train_test_split(root_ins_df,
                                     stratify=root_ins_df['policies_sold'],
                                     test_size=0.3,
                                     random_state=43)
skf = StratifiedKFold(n_splits=5)
```

```python
baseline_profit = []
profit_array = []
plt.figure(figsize=(14,8))
legend_added = False
for i in range(4):
    for train_index, test_index in skf.split(train_data.loc[:,train_data.columns != 'policies_sold'],
                                             train_data['policies_sold']):

        bidder_inst = naive_bidder(tol=0.0,reinforce=True,fixed_reward=40)
        bidder_inst.train(train_data.iloc[train_index],train_price_rank=False)
        baseline_reward, rewards_array = bidder_inst.test(train_data.iloc[test_index])
        baseline_profit += [sum(baseline_reward)]
        profit_array += [sum(rewards_array)]
        if(legend_added):
            plt.plot(np.cumsum(baseline_reward),'tab:orange',alpha=0.8)
            plt.plot(np.cumsum(rewards_array),'tab:blue',alpha=0.8)
        else:
            plt.plot(np.cumsum(baseline_reward),'tab:orange',label=f'baseline',alpha=0.8)
            plt.plot(np.cumsum(rewards_array),'tab:blue',label='naive bidding',alpha=0.8)
            legend_added = True
plt.title('Net Reward vs Time for 20 runs with a Fixed Reward Per Successful Bid of $40 and No Reinforcement')
plt.ylabel('Reward')
plt.xlabel('Time')
plt.legend()
```

```python
data1 = baseline_profit
data2 = profit_array
width =0.4
plt.figure(figsize=(10,6))
plt.bar(np.arange(len(data1)), data1, width=width,label='Baseline')
plt.bar(np.arange(len(data2))+ width, data2, width=width,label='Naive bidding')
plt.xlabel('Run')
plt.ylabel('Total Reward')
plt.title('Total Reward Comparision between Constant $10 Bidding and Naive Bidding')
plt.legend()
plt.show()
# plt.plot(baseline_profit,'tab:orange',label=f'baseline',alpha=0.8)
# plt.plot(profit_array)
```

```python
reinforcement_profit = []
no_reinforcement_profit = []
plt.figure(figsize=(14,8))
legend_added = False
for i in range(0,10):
    for train_index, test_index in skf.split(train_data.loc[:,train_data.columns != 'policies_sold'],
                                             train_data['policies_sold']):

        bidder_inst = naive_bidder(tol=0.0,reinforce=True,fixed_reward=40)
        bidder_inst_2 = naive_bidder(tol=0.0,reinforce=False,fixed_reward=40)
        bidder_inst.train(train_data.iloc[train_index])
        bidder_inst_2.train(train_data.iloc[train_index])
        baseline_reward_1, reinforced_profit_i = bidder_inst.test(train_data.iloc[test_index])
        baseline_reward_2, no_reinforced_profit_i = bidder_inst_2.test(train_data.iloc[test_index])
    reinforcement_profit += [sum(reinforced_profit_i)/sum(baseline_reward_1)]
    no_reinforcement_profit += [sum(no_reinforced_profit_i)/sum(baseline_reward_2)]
#         if(legend_added):
#             plt.plot(np.cumsum(baseline_reward),'tab:orange',alpha=0.8)
#             plt.plot(np.cumsum(rewards_array),'tab:blue',alpha=0.8)
#         else:
#             plt.plot(np.cumsum(baseline_reward),'tab:orange',label=f'baseline',alpha=0.8)
#             plt.plot(np.cumsum(rewards_array),'tab:blue',label='naive bidding',alpha=0.8)
#             legend_added = True
# plt.title('Net Reward vs Time for 20 runs with a Fixed Reward Per Successful Bid of $40 and No Reinforcement')
# plt.ylabel('Reward')
# plt.xlabel('Time')
# plt.legend()


plt.figure(figsize=(10,6))
plt.plot(reinforcement_profit,'tab:orange',alpha=0.8,label='Reinforcement Naive Bidding')
plt.plot(no_reinforcement_profit,'tab:blue',alpha=0.8,label='Non-Reinforcement Naive Bidding')
# plt.bar(np.arange(len(data1)), data1, width=width,label='Baseline')
# plt.bar(np.arange(len(data2))+ width, data2, width=width,label='Naive bidding')
plt.xlabel('Run')
plt.ylabel('Total Reward')
plt.title('Total Reward Comparision between Reinforced and Non-Reinforced Naive Bidding')
plt.legend()
plt.show()
```
